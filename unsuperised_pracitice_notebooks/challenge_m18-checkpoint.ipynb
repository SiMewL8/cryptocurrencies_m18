```python
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
import plotly.express as px
import hvplot.pandas
```

```python
file_path = "crypto_data.csv"
crypto_df = pd.read_csv(file_path)
crypto_df.head()
crypto_df.dtypes
```

```python
crypto_df
```

```python
# Remove all cryptocurrencies that arent trading.
crypto_df = crypto_df[crypto_df['IsTrading'] != 0]
crypto_df
```

```python
for col in crypto_df.columns:
    print(f"{col} has {crypto_df[col].isnull().sum()} null values")
```

```python
# Remove all cryptocurrencies that dont have an algorithm defined
for r in crypto_df.Algorithm.unique():
    print(r)
```

```python
for r in crypto_df.TotalCoinSupply.unique():
    print(r)
```

```python
crypto_df['TotalCoinSupply'] = pd.to_numeric(crypto_df['TotalCoinSupply'], errors='coerce')
crypto_df
```

```python
# Remove the IsTrading column
crypto_df = crypto_df.drop(['IsTrading'], axis=1)
crypto_df = crypto_df.set_index(['Unnamed: 0'])
crypto_df
```

```python
# Remove all cryptocurrencies without coins mined.
crypto_df = crypto_df.dropna(subset=['TotalCoinsMined'])
crypto_df = crypto_df[crypto_df['TotalCoinsMined'] != 0]
crypto_df
```

```python
# Store the names of all cryptocurrencies on a DataFramed named coins_name,
# and use the crypto_df.index as the index for this new df

coins_name = pd.DataFrame(crypto_df['CoinName'])
coins_name
```

```python
# remove coiname column
crypto_df = crypto_df.drop(['CoinName'],axis=1)
crypto_df.dtypes
```

```python
# Create dummies variables for all of the text features,
# and store the resulting data on a DataFrame named X.

X = pd.get_dummies(data=crypto_df, columns=['Algorithm', 'ProofType'])
X.info()
```

```python
#standardscaler for crypto
from sklearn.preprocessing import StandardScaler
scale_crypto = StandardScaler().fit_transform(X)
```

```python
#PCA components = 3
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
crp_pca = pca.fit_transform(scale_crypto)
pcs_df = pd.DataFrame(
    data=crp_pca,
    columns=['PC 1', 'PC 2', 'PC 3'],
    index=crypto_df.index
)

pcs_df.head()
```

```python
# Find the best value for K
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)

# Create the elbow curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")

#k=6 looks like the best fit
```

```python
# Initialize the K-means model
model = KMeans(n_clusters=3, random_state=0)

# Fit the model
model.fit(pcs_df)

# Predict clusters
predictions = model.predict(pcs_df)

# Add the predicted class columns
pcs_df["class"] = model.labels_
pcs_df.head()
```

```python
#new df clustered_df
clustered1_df = pd.merge(crypto_df,pcs_df, on='Unnamed: 0')
# clustered1_df.head()

clustered_df = pd.merge(clustered1_df, coins_name, on='Unnamed: 0')
clustered_df.head()
```

```python
fig = px.scatter_3d(
    clustered_df,
    x="PC 1",
    y="PC 2",
    z="PC 3",
    color="class",
    symbol="class",
    width=800,
    hover_name="CoinName",
    hover_data=["Algorithm"]
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```

```python
hov_df = clustered_df.hvplot.table(columns=[
    'CoinName',
    'Algorithm',
    'ProofType',
    'TotalCoinSupply',
    'TotalCoinsMined',
    'class'], 
    sortable=True,
    selectable=True)
```
